```sql
%python
df = spark.read.csv(
    "/Volumes/log_file/log_data/timings",
    header=True,
    inferSchema=True
)
df.show()
```

```sql
%python
df.createOrReplaceTempView("LoginHistory")
```

```sql
%sql

SELECT *
FROM LoginHistory;
```

```sql
%sql

SELECT
    UserID,
    LoginTime,
    LAG(LoginTime)
    OVER
    (
        PARTITION BY UserID
        ORDER BY LoginTime
    ) AS PreviousLogin
FROM LoginHistory;
```

```sql
WITH Ranked AS
(
    SELECT *,
           ROW_NUMBER() OVER
           (
               PARTITION BY DeptID
               ORDER BY Salary DESC
           ) AS rn
    FROM Employee
)

SELECT *
FROM Ranked
WHERE rn = 1;
```

```sql
WITH Ranked AS
(
    SELECT *,
           ROW_NUMBER() OVER
           (
               PARTITION BY DeptID
               ORDER BY Salary DESC
           ) AS rn
    FROM Employee
)

SELECT *
FROM Ranked
WHERE rn <= 3;
```
