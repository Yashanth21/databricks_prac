```python
file_path = "/Volumes/employee/emp/empingestion/employee.csv"

employees = (
    spark.read
    .option("header", True)
    .csv(file_path)
)

print("Employee Data")
display(employees)

employees.show(5)

print("Columns:")
print(employees.columns)

print("Schema:")
employees.printSchema()

print("Total Records:", employees.count())

print("Employee Names and Departments")
display(employees.select("name", "department"))

print("IT Employees")
display(employees.filter(employees.department == "IT"))
```
