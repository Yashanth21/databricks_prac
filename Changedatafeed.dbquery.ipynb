```sql
CREATE TABLE Employeee
(
    EmpID INT,
    EmpName STRING,
    Salary INT
)
USING DELTA
TBLPROPERTIES
(
    delta.enableChangeDataFeed = true
);

INSERT INTO Employeee
VALUES
(1,'John',50000),
(2, 'Mary',60000);

UPDATE Employeee
SET Salary = 65000
WHERE EmpID = 2;

DELETE FROM Employeee
WHERE EmpID = 1;

SELECT *
FROM table_changes('Employeee',1);

SELECT *
FROM table_changes('Employeee', 2);

DESCRIBE HISTORY Employeee;

SELECT *
FROM table_changes('Employeee', 3);
```
