```python
import csv
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s - %(message)s"
)

file_path = "/Volumes/employee/emp/empingestion/employee.csv"

employees = []

logging.info("Starting employee ingestion...")

with open(file_path, "r", encoding="utf-8") as file:

    reader = csv.DictReader(file)

    for row in reader:

        if row["name"] == "":
            logging.warning(
                "Skipping Employee %s because name is missing.",
                row["id"],
            )
            continue

        try:
            row["salary"] = int(row["salary"])
        except ValueError:
            logging.warning(
                "Skipping Employee %s because salary is invalid.",
                row["id"],
            )
            continue

        employees.append(row)

logging.info("Successfully loaded %d valid employees.", len(employees))

for employee in employees:
    logging.info(employee)
```
